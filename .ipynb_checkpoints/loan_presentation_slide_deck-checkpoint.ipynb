## Dataset Overview

> This dataset is on Loan and their outcomes. The loan disbursing organization is Prosper Loan organizayion. This data set contains 113,937 loans with 81 variables on each loan, including loan status, borrower APR (or interest rate), available bank card credit, employment status duration, listing category, borrower monthly income, and many others. The data cover loans from  quarter 1 2007 to quarter 1 2014.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style("darkgrid")

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_prosper = pd.read_csv("prosper_extract_adjusted.csv")
```

## Borrower's Monthly Income Determines if they will complete their loans.

> Borrowers that fall into the very_high income category have high bank card credit and higher rate of loan completion. Those with low income levels have less-favourable loan outcome.

```python
# Visualizing the relationship of Loan Status to Available Bank Card Credit and Income category
plt.figure(figsize = (15,8))
ax = sb.barplot(data = df_prosper, x = "income_category", y = "AvailableBankcardCredit", hue = "LoanStatus")
ax.legend(loc = 2, ncol = 6, framealpha = 1, title = "LoanStatus")
plt.title("How Income and Bank Balance affects Loan Status I");
```

```python
# Visualizing the relationship of Loan Status to Available Bank Card Credit and Income category
plt.figure(figsize = (15,8))
ax = sb.pointplot(data = df_prosper, x = "income_category", y = "AvailableBankcardCredit", hue = "LoanStatus", dodge = 0.3, linestyles = "")
ax.legend(loc = 2, ncol = 6, framealpha = 1, title = "LoanStatus")
plt.title("How Income and Bank Balance affects Loan Status II");
```

## Borrowers in Full time employment PAY BACK LOANS on TIME

> Those borrowers who hold full-time employment fares better in paying back their loans. They have lower rates of defaults.

```python
# Visualizing the relationship between variables and Loan Status
plt.figure(figsize=(10,6))
sb.countplot(data = df_prosper, x = "EmploymentStatus", hue = "LoanStatus")
plt.legend(loc = 1, ncol = 2, framealpha = 1, title = "Loan Status")
plt.ylabel("Number of Borrowers")
plt.xlabel("Employment Status of Borrowers")
plt.title("Does Employment Status affect Loan Status?");
```

## Short and Long Term loans are not easily recovered.
#### Favour medium term loans

> Borrowers find it difficult to complete loans whose terms are above 36 months. Generally, loan terms of 36 months have better completion rate.

```python
# Visualizing term_category and LoanStatus
plt.figure(figsize=(20,6))
sb.countplot(data = df_prosper, x = "term_category", hue = "LoanStatus")
plt.legend(loc = 1, ncol = 2, framealpha = 1, title = "Loan Status")
plt.ylabel("Number of Borrowers")
plt.xlabel("Category of Loan based on Term")
plt.title("Does Loan Term affects Loan Status?");
```

```python
# Visualizing the relationship of term to Available Bank Card Credit and Stated Monthly Income
g=sb.FacetGrid(data = df_prosper, aspect=1.2, height=5, col='Term', col_wrap=4)
g.map(sb.regplot, 'StatedMonthlyIncome', 'AvailableBankcardCredit', x_jitter=0.04, scatter_kws={'alpha':0.1});
g.add_legend();
```

## Insist on Recommendations
> Borrowers who have at least one recommendation have a higher rate of completion than those who don't.

> If such borrower has a high bank credit and a high monthly income, I can bet he/she will complete the loan even if the employment status duration is not so high

```python
# Visualization of income, card credit, recommendations, and employment duration
fig, ax = plt.subplots(ncols=2, figsize=[12,6])
sb.pointplot(data = df_prosper, x = 'EmploymentStatusDuration', y = 'StatedMonthlyIncome', hue = 'Recommendations',
           palette = 'Purples', linestyles = '', dodge = 0.4, ax=ax[0])
sb.pointplot(data = df_prosper, x = 'EmploymentStatusDuration', y = 'AvailableBankcardCredit', hue = 'Recommendations',
           palette = 'Greens', linestyles = '', dodge = 0.4, ax=ax[1]);
plt.title("Recommendations, Employment Status Duration, Bank card Credit, and Monthly Income")
```

## What should I do?
> Now you know the why, then do these:
- Insist that borrowers be employed
- Check their monthly income and available bank card credit.
- Let them have at least one recommendation.
- Extend more short term loans of 36 months duration, and 
- Extend more loans to borrowers in the lower listing categories.
